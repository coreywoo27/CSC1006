Q10 (a)
calculate MSE when w = 1.2 and b = 0.5
(I have the result rounded to 3 significant numbers)

```python
l1=np.array([1.0,2.0,3.0])
l2=np.array([2.0,3.5,4.5])
MSE=lambda w,b: np.mean((l2-(w*l1+b))**2)
print(MSE(1.2,0.5).round(3))

```
